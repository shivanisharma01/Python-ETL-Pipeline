```python
def create_engine_sql():
    return create_engine(
        "mssql+pyodbc://@SQL2K16FCL541.omega.dce-eir.net\SQL2016/SOURCE?driver=ODBC+Driver+17+for+SQL+Server&Encrypt=no&TrustServerCertificate=yes&connect_timeout=60&QueryTimeout=60",
        fast_executemany=True
    )

def get_chunknum(df):
    df_len = len(df.columns)
    return math.floor(2100/df_len)

def optimize_dataframe(df):
    for col in df.select_dtypes(include=['float']):
        df[col] = pd.to_numeric(df[col], downcast='float')
    for col in df.select_dtypes(include=['int']):
        df[col] = pd.to_numeric(df[col], downcast='integer')
    return df

def batch_insert_to_sql(df_chunk, table_name):
    df_chunk = optimize_dataframe(df_chunk)
    engine_sql = create_engine_sql()  # Move connection setup inside worker
    chunknum = get_chunknum(df_chunk)
    
    with engine_sql.connect() as conn:
        df_chunk.to_sql(table_name, conn, schema='DTC', if_exists='append', index=False, method='multi', chunksize=chunknum)

def process_chunk(df_chunk, table_name):
    try:
        # Insert data chunk into SQL Server
        batch_insert_to_sql(df_chunk, table_name)
        print(f"Chunk inserted successfully!")
    except Exception as e:
        print(f"Failed to insert chunk: {e}")        
        
def parallel_etl(df_iterator, table_name, max_workers=4):
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = []
        
        for df_chunk in df_iterator:
            # Submit each chunk to be processed in parallel
            futures.append(executor.submit(process_chunk, df_chunk, table_name))
        
        # Wait for all futures to complete
        for future in as_completed(futures):
            try:
                future.result()  # This will raise any exceptions caught during execution
            except Exception as e:
                print(f"Error occurred during chunk processing: {e}")

parallel_etl(df_iterator, 'TT1PROV', max_workers=4)

print("ETL job completed successfully")
```
